Una empresa dedicada a la comercialización de un producto de mucho éxito quiere determinar el
número de unidades que debe comprar con el objetivo de satisfacer la demanda de sus clientes los próximos meses. La distribución de las ventas de los últimos tres años se recoge en la tabla siguiente.

[Ver tabla del pizarrón]

El coste de cada unidad de producto asciende a 750 euros, siendo su precio de venta unitario de 1000 euros. Se sabe que dentro de un mes saldrá al mercado un nuevo modelo de características muy superiores al modelo actual. Cuando esto ocurra la empresa podrá devolver al distribuidor las unidades
sobrantes del modelo actual y este le indemnizará con 250 euros por cada unidad retornada.

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
precio_venta = 1000
precio_dev = 250
precio_compra = 750
def Beneficio(ventas, compras = 400):
  if ventas < compras:
    beneficio = ventas * precio_venta +(compras - ventas) * precio_dev - compras * precio_compra
  else:
    beneficio = compras * precio_venta - compras * precio_compra
  return beneficio
```

```python
Beneficio(500)
```

```python
def Venta():
  venta = r.random()
  if 0 <= venta < 0.22:
    U_ven = 100
  elif 0.22 <= venta < 0.47:
    U_ven = 200
  elif 0.47 <= venta < 0.58:
    U_ven = 300
  elif 0.58 <= venta < 0.83:
    U_ven = 400
  elif 0.83 <= venta < 0.94:
    U_ven = 500
  elif 0.94 <= venta <= 1:
    U_ven = 600
  return U_ven
```

```python
Venta()
```

```python
def Beneficio_prom (n):
  beneficios = []
  promedio = []
  for i in range(n):
    ventas = Venta()
    beneficios = Beneficio(ventas)
    p = np.sum(beneficios)/(i + 1)
    promedio.append(p)
  return promedio
```

```python
l=Beneficio_prom(10)
```
